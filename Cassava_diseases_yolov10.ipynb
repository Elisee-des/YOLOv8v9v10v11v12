```python
from ultralytics import YOLO

# Charger le modèle YOLOv10n (version la plus légère)
model = YOLO('yolov10n.pt')

# Commencer l'entraînement
model.train(
    data='/content/drive/MyDrive/Maching_learning/Cassava_diseases_detection.v2i.yolov5pytorch/data.yaml',
    epochs=30,
    batch=8,
    imgsz=640,
    project='/content/drive/MyDrive/Maching_learning/Yolov10/runs/results/',  # Sauvegarde sur Drive
    name='yolov10n_train'
)
```
